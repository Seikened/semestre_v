```python
#%pip install yfinance
#%pip install seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
```

```python
# Bajamos los datos desde que existen datos hasta hoy

data = yf.download("BTC-USD", start="2024-01-01", end="2025-09-05")
```

```python
data['Retorno'] = data['Close'].pct_change() # pct -> Genera calculo completo

data['LogRetorno'] = np.log(data['Close'] / data['Close'].shift(1))

plt.plot(data['Retorno'], label="Rendimiento Bitcoin")
plt.plot(data['LogRetorno'], label="Rendimiento Logarítmico Bitcoin")
plt.title("Rendimiento de Bitcoin")
plt.xlabel("Fecha")
plt.ylabel("USD")
plt.grid(True)
plt.legend()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(data['LogRetorno'].dropna(), bins=50, kde=True,color='green')
sns.histplot(data['Retorno'].dropna(), bins=50, kde=True,color='blue')
plt.title("Distribución de Rendimientos de Bitcoin")
plt.xlabel("Rendimiento")
plt.grid(True)
```
