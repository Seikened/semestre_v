```python
import numpy as np
import pandas as pd


data = pd.read_csv("case1_tecnologia_growth.csv")

if data['return'].abs().max() > 1:
    data['return'] = data['return'] / 100


rendimiento_anual = 0.05
dias = 252
rendimiento_diario = (1 + rendimiento_anual)**(1/dias) - 1

promedio_diario = data['return'].mean()                       
ruido_diario = data['return'].std(ddof=1)                   
volumen_anual = ruido_diario * np.sqrt(dias)
sharpe_anual = ((promedio_diario - rendimiento_diario) / ruido_diario) * np.sqrt(dias)

downside_diff = np.minimum(0, data['return'] - rendimiento_diario)
downside_deviation = np.sqrt(np.mean(downside_diff**2))

sortino_anual = ((promedio_diario - rendimiento_diario) / downside_deviation) * np.sqrt(dias)

print(f"Media diaria: {promedio_diario:.6f}")
print(f"Volatilidad diaria: {ruido_diario:.6f}")
print(f"Volatilidad anualizada: {volumen_anual:.6f}")
print(f"Sharpe anualizado: {sharpe_anual:.6f}")
print(f"Desviación a la baja (Downside Deviation): {downside_deviation:.6f}")
print(f"Sortino anualizado: {sortino_anual:.6f}")


if sharpe_anual > 1:
    print("\nInterpretación: El portafolio compensa bien el riesgo asumido.")
elif sharpe_anual > 0:
    print("\nInterpretación: El rendimiento compensa moderadamente el riesgo.")
else:
    print("\nInterpretación: El rendimiento NO compensa el riesgo asumido.")
```
