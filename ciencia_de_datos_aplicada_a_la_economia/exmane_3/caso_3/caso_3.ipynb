```python
# caso3_prediccion_regresion.py
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


path = "case1_tecnologia_growth.csv"
data = pd.read_csv(path)

if data['return'].abs().max() > 1:
    data['return'] = data['return'] / 100.0

data['t'] = np.arange(1, len(data) + 1)


X = data[['t']]
y = data['return']   


modelop = LinearRegression()
modelop.fit(X, y)


b0 = modelop.intercept_
b1 = modelop.coef_[0]


r2 = r2_score(y, modelop.predict(X))


t = np.array([[61]])
y_pred = modelop.predict(t)[0]


print("=== Caso 3 â€” PredicciÃ³n de Rendimientos con RegresiÃ³n Lineal ===")
print(f"Pendiente (Î²â‚): {b1:.6f}")
print(f"Intercepto (Î²â‚€): {b0:.6f}")
print(f"Coeficiente de determinaciÃ³n (RÂ²): {r2:.6f}")
print(f"PredicciÃ³n del rendimiento para el dÃ­a 61: {y_pred:.6f}")


if b1 > 0:
    tendencia = "alcista ğŸ“ˆ"
elif b1 < 0:
    tendencia = "bajista ğŸ“‰"
else:
    tendencia = "neutral (sin tendencia clara)"

print(f"\nInterpretaciÃ³n:")
print(f"- La pendiente es {b1:.6f}, lo que indica una tendencia {tendencia}.")
if r2 < 0.2:
    print("- El RÂ² es bajo: la relaciÃ³n entre el tiempo y los rendimientos es dÃ©bil.")
elif r2 < 0.5:
    print("- El RÂ² es moderado: existe cierta tendencia, pero con bastante ruido.")
else:
    print("- El RÂ² es alto: el tiempo explica bien la variaciÃ³n de los rendimientos.")
```
